```python
%matplotlib inline
from pycocotools.coco import COCO
import numpy as np
import skimage.io as io
import matplotlib.pyplot as plt
import pylab
import os
import shutil
pylab.rcParams['figure.figsize'] = (8.0, 10.0)
```

```python
dataDir='..'
dataType='val2017'
annFile='{}/annotations/instances_{}.json'.format(dataDir,dataType)
```

```python
import json

with open(annFile, "r") as f:
    annjs = json.load(f)
annjs.keys()
```

```python
# initialize COCO api for instance annotations
coco=COCO(annFile)
```

```python
# get all images containing given categories, select one at random
catIds = coco.getCatIds(catNms=['person',]);
imgIds = coco.getImgIds(catIds=catIds, );
# imgIds = coco.getImgIds(imgIds = [324158])
img = coco.loadImgs(imgIds[np.random.randint(0,len(imgIds))])[0]
```

```python
annjs["info"]["description"] = 'Only person from COCO 2017 Dataset'
annjs["images"] = coco.loadImgs(imgIds)
annjs["annotations"] = coco.loadAnns(coco.getAnnIds(imgIds, iscrowd=0))
annjs["categories"] = [{'supercategory': 'person', 'id': 1, 'name': 'person'}]
```

```python
outDataType = "person_val2017"
with open('{}/annotations/instances_{}.json'.format(dataDir,outDataType), "w") as f:
    json.dump(annjs, f)
```

```python
sourceDir = '../val2017'
saveDir = '../person_val2017'

os.makedirs(saveDir, exist_ok=True)
for img in coco.loadImgs(imgIds):
    fname = str(img["id"]).zfill(12) + ".jpg"
    shutil.copy(os.path.join(sourceDir, fname), saveDir)
```

```python
# load and display image
# I = io.imread('%s/images/%s/%s'%(dataDir,dataType,img['file_name']))
# use url to load image
I = io.imread(img['coco_url'])
plt.axis('off')
plt.imshow(I)
plt.show()
```

```python
# load and display instance annotations
plt.imshow(I); plt.axis('off')
annIds = coco.getAnnIds(imgIds=img['id'], catIds=catIds, iscrowd=1)
anns = coco.loadAnns(annIds)
coco.showAnns(anns)
```

```python
# initialize COCO api for person keypoints annotations
annFile = '{}/annotations/person_keypoints_{}.json'.format(dataDir,dataType)
coco_kps=COCO(annFile)
```

```python
# load and display keypoints annotations
plt.imshow(I); plt.axis('off')
ax = plt.gca()
annIds = coco_kps.getAnnIds(imgIds=img['id'], catIds=catIds, iscrowd=None)
anns = coco_kps.loadAnns(annIds)
coco_kps.showAnns(anns)
```

```python
# initialize COCO api for caption annotations
annFile = '{}/annotations/captions_{}.json'.format(dataDir,dataType)
coco_caps=COCO(annFile)
```

```python
# load and display caption annotations
annIds = coco_caps.getAnnIds(imgIds=img['id']);
anns = coco_caps.loadAnns(annIds)
coco_caps.showAnns(anns)
plt.imshow(I); plt.axis('off'); plt.show()
```
